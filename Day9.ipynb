```python
# Step 1: Define user-defined functions
def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y == 0:
        return "Error: Division by zero is not allowed."
    else:
        return x / y

```

```python
# Step 2: Ask the user for input

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
operation = input("Choose operation (+, -, *, /): ")


```

```python
# Step 3: Perform calculation using chosen function
if operation == "+":
    result = add(num1, num2)
elif operation == "-":
    result = subtract(num1, num2)
elif operation == "*":
    result = multiply(num1, num2)
elif operation == "/":
    result = divide(num1, num2)
else:
    result = "Invalid operation selected."

```

```python

# Step 4: Use built-in functions (abs, round, pow)
if isinstance(result, (int, float)):
    print("\nRaw Result:", result)
    print("Absolute Value:", abs(result))
    print("Rounded (2 decimals):", round(result, 2))
    print("Result squared using pow():", pow(result, 2))
else:
    print("\n", result)

```
