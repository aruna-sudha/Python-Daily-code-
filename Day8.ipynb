```python
# Project Title: Employee Hours Report (Exam Scores Analysis)
import random

```

```python
# Step 1:Create a list of 10 random exam scores (0â€“100) using list comprehension
scores = [random.randint(0, 100) for i in range(10)]


```

```python
# Step 2: Display each score with its index using enumerate
print("Exam Scores with Index:")
for index, score in enumerate(scores):
    print(f"Index {index}: Score = {score}")

```

```python
# Step 3: Find all scores greater than 75 and display them
high_scores = [s for s in scores if s > 75]
print("\nScores greater than 75:")
print(high_scores)

```

```python

# Step 4: Calculate and display the average score
average_score = sum(scores) / len(scores)
print("\nAverage Score:", round(average_score, 2))

```
