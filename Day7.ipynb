```python
# Project Title: Number Guessing Game
import random

# Step 1: Randomly choose a number between 1 and 20


secret_number = random.randint(1, 20)

attempts = 0
max_attempts = 5

```

```python
# Step 2: While loop for guessing
while attempts < max_attempts:
    try:
        guess = int(input(f"Attempt {attempts+1}: Guess the number (1-20): "))
    except ValueError:
        print("Invalid input! Please enter an integer.")
        continue   # skip this attempt without counting

    attempts += 1

    if guess == secret_number:
        print("Correct! You guessed the number!")
        break
    elif guess < secret_number:
        print("Too low! Try again.")
    else:
        print("Too high! Try again.")

```

```python
#Step 3: This block executes if loop finishes without break
print(f"Sorry, you've used all {max_attempts} attempts.")
print(f"The correct number was {secret_number}.")
```
