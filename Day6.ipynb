```python
# Project Title: BMI Calculator
# Step 1: Ask the user for weight and height

weight = float(input("Enter your weight in kilograms: "))
height = float(input("Enter your height in meters: "))

```

```python

# Step 2: Ensure positive values
if weight <= 0 or height <= 0:
    print("Error: Weight and height must be positive numbers.")
else:
    # Step 3: Calculate BMI
    bmi = weight / (height * height)

```

```python
# Step 4: Classify BMI using conditional statements
if bmi < 18.5:
        category = "Underweight"
elif bmi < 25:
        category = "Normal"
elif bmi < 30:
        category = "Overweight"
else:
        category = "Obese"

```

```python
   # Step 5: Display the result clearly
print("\nYour BMI is:", round(bmi, 2))
print("Health Category:", category)

```
